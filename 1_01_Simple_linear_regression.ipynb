```python
import pandas as pd
url = '1.01. Simple linear regression.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['SAT']), y = data['GPA'])
print(model.intercept_)
print(model.coef_)
```

```python
data.plot(kind = 'scatter', x= 'SAT', y='GPA', figsize = (9,9), color= 'black', xlim = (1500,2200))
y_pred = model.predict(X = pd.DataFrame(data['SAT']))
plt.plot(data['SAT'],y_pred,color = 'blue')
model.score(X = pd.DataFrame(data['SAT']), y = data['GPA'])
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['SAT'],data['SAT']**2]).T # tao ham y=
model.fit(X, y= data['GPA'])
print('Intercept của mô hình', model.intercept_)
print('Coeffident của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y= data['GPA']))
import numpy as np
range = np.arange(1500,2200,0.1)
data.plot(kind = 'scatter', x = 'SAT', y = 'GPA', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2]).T  #**2 là bậc 2
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color ='blue')
```
