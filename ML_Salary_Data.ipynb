```python
import pandas as pd
url = 'Salary_Data.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['Salary']), y = data['YearsExperience'])
print(model.intercept_)
print(model.coef_)
```

```python
data.plot(kind = 'scatter', x= 'Salary', y='YearsExperience', figsize = (9,9), color= 'black', xlim = (0,150000))
y_pred = model.predict(X = pd.DataFrame(data['Salary']))
plt.plot(data['Salary'],y_pred,color = 'blue')
model.score(X = pd.DataFrame(data['Salary']), y = data['YearsExperience'])
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['Salary'],data['Salary']**2]).T # tao ham y=
model.fit(X, y= data['YearsExperience'])
print('Intercept của mô hình', model.intercept_)
print('Coeffident của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y= data['YearsExperience']))
```

```python
import numpy as np
range = np.arange(30000,150000,0.1)
data.plot(kind = 'scatter', x = 'Salary', y = 'YearsExperience', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2]).T  #**2 là bậc 2
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color ='blue')
```
