```python
import pandas as pd
url = '50_Startups.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show
```

```python
correlation = data.corr()
print(correlation)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['R&D Spend']), y = data['Profit'])
print(model.intercept_)
print(model.coef_)
```

```python
data.plot(kind = 'scatter', x= 'R&D Spend', y='Profit', figsize = (9,9), color= 'black', xlim = (0,150000))
y_pred = model.predict(X = pd.DataFrame(data['R&D Spend']))
plt.plot(data['R&D Spend'],y_pred,color = 'blue')
model.score(X = pd.DataFrame(data['R&D Spend']), y = data['Profit'])
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'],data['Profit']**2]).T # tao ham y=
model.fit(X, y= data['Profit'])
print('Intercept của mô hình', model.intercept_)
print('Coeffident của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y= data['Profit']))
import numpy as np
range = np.arange(0,150000,10)
data.plot(kind = 'scatter', x = 'R&D Spend', y = 'Profit', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2]).T  #**2 là bậc 2
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color ='blue')
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['R&D Spend'],data['Profit']**2,data['Profit']**3]).T # tao ham y=
model.fit(X, y= data['Profit'])
print('Intercept của mô hình', model.intercept_)
print('Coeffident của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y= data['Profit']))
import numpy as np
range = np.arange(0,150000,10)
data.plot(kind = 'scatter', x = 'R&D Spend', y = 'Profit', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2,range**3]).T  #**2 là bậc 2
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color ='blue')
```
