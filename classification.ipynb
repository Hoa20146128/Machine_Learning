```python
import pandas as pd
url = 'classification.csv'
data = pd.read_csv(url)
import matplotlib.pyplot as plt
data.hist()
plt.show
correlation = data.corr()
print(correlation)
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
model.fit(X = pd.DataFrame(data['success']), y = data['interest'])
print(model.intercept_)
print(model.coef_)
```

```python
data.plot(kind = 'scatter', x= 'success', y='interest', figsize = (9,9), color= 'black', xlim = (0,1.5))
y_pred = model.predict(X = pd.DataFrame(data['success']))
plt.plot(data['success'],y_pred,color = 'blue')
model.score(X = pd.DataFrame(data['success']), y = data['interest'])
```

```python
model = linear_model.LinearRegression()
X = pd.DataFrame([data['success'],data['success']**2]).T # tao ham y=
model.fit(X, y= data['interest'])
print('Intercept của mô hình', model.intercept_)
print('Coeffident của mô hình:', model.coef_)
print('Độ chính xác của mô hình:', model.score(X, y= data['interest']))
import numpy as np
range = np.arange(0,1.5,0.1)
data.plot(kind = 'scatter', x = 'success', y = 'success', figsize = (9,9), color = 'black')
predictor = pd.DataFrame([range,range**2]).T  #**2 là bậc 2
y_pred = model.predict(X = predictor)
plt.plot(range, y_pred, color ='blue')
```
